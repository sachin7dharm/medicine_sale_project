This dataset is downloded from Kaggle https://www.kaggle.com/datasets/shivamkapoor5034/pharmaceutical-monthly-sales?select=MedicineData.csv

The dataset contains monthly sales of the drug R03 in the United States (in millions of US $) between 1992 and 2018.
R03 - Drugs for obstructive airway diseases.
Sales data are resampled to the monthly periods. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# coverting date column into date dat type

df['date'] = pd.to_datetime(df['date'])
df
```

```python
#creating new column named year

df['year'] = df['date'].dt.year
df['month']=df['date'].dt.month_name()
df
```

```python
# year wise sale

ydf=df.groupby('year')['R03'].sum().reset_index()
sns.lineplot(x='year',y='R03',data=ydf,marker='o')
plt.grid()
plt.title('Year wise Sale')

plt.show()
```

```python
# year wise average sale

avg_ydf=df.groupby('year')['R03'].mean().reset_index()
sns.lineplot(x='year',y='R03',data=avg_ydf,marker='o')
plt.grid()
plt.title('Year wise Average Sale')

plt.show()
```

```python
# month wise total sale from 1992 to 2018

mdf=df.groupby(['month'])['R03'].sum().sort_values().reset_index()
sns.barplot(x='R03',y='month',data=mdf)
plt.show()
```

```python
# year and month wise sale

ymdf=df.groupby(['year','month'])['R03'].sum().reset_index()
sns.lineplot(x='year',y='R03',hue='month',data=ymdf)
plt.grid()
plt.show()
```
